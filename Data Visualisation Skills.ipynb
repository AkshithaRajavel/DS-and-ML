B. Data Visualization using Matplotlib for IRIS Dataset:

a. For IRIS dataset create a scatter plot in Matplotlib, also create a figure and an axis
using plt.subplots so we can give our plot a title and labels.

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(20)
f.suptitle("iris_dataset")
j=1
for i in iris_dataset.columns[:-1]:
  plt.subplot(2,2,j)
  plt.scatter(iris_dataset[i],iris_dataset["class"],c='b')
  plt.xlabel(i)
  plt.ylabel("class")
  j+=1
```

b. Now give the graph more meaning by colouring in each data-point by its class. This can
be done by creating a dictionary which maps from class to colour and then scattering
each point on its own using a for-loop and passing the respective colour.

```python
class_colour = {}

for i in range(len(iris_dataset.columns[:-1])):
  class_colour[iris_dataset.columns[i]]=list(mcolors.BASE_COLORS.keys())[i]
class_colour


for i in iris_dataset.columns[:-1]:
  plt.scatter(iris_dataset[i],iris_dataset["class"],c=class_colour[i])
plt.legend(class_colour.keys())
```

c. Now create a line chart for IRIS Dataset by calling the plot method. Also plot multiple
columns in one graph, by looping through the columns we want and plotting each
column on the same axis.

```python
for i in iris_dataset.columns[:-1]:
  plt.plot(iris_dataset[i],iris_dataset["class"],c=class_colour[i])
plt.legend(class_colour.keys())
```

C. Data Visualization using Matplotlib for Wine Dataset:

a. For Wine Dataset create a Histogram using the hist method. If we pass it categorical
data like the points column from the wine dataset it will automatically calculate how often
each class occurs.

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
f.suptitle("wine_dataset")
j=1
col=["country","variety"]
for i in col:
    plt.subplot(2,1,j)
    plt.title(i)
    plt.hist(wine_dataset.iloc[:10][i])
    j+=1
```

b. Use the Bar method to create a bar chart. A bar chart can be created using the bar
method. The bar chart isn’t automatically calculating the frequency of a category so we
are going to use pandas value_counts function to do this.

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
j=1
for i in col:
  names=wine_dataset.iloc[:10][i].value_counts().index
  values=wine_dataset.iloc[:10][i].value_counts().values
  plt.subplot(2,1,j)
  plt.bar(names,values)
  plt.title(i)
  j+=1

#wine_dataset.iloc[:10]["variety"].value_counts().index
```

a. Import seaborn library.

```python
import seaborn as sns
```

b. For IRIS dataset use scatterplots and line plot method from Seaborn for creating a
scatterplot and line plot

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
j=1
for i in iris_dataset.columns[:-1]:
  plt.subplot(2,2,j)
  sns.scatterplot(data=iris_dataset,x=i,y="class")
  j+=1
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
j=1
for i in iris_dataset.columns[:-1]:
  plt.subplot(2,2,j)
  sns.lineplot(data=iris_dataset,x=i,y="class")
  j+=1
```

c. For Wine Dataset create a histogram in Seaborn using sns.distplot method. Now create
a bar chart in Seaborn using the sns.countplot method for Wine Dataset.

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
j=1
col=["price","points"]
for i in col:
  plt.subplot(2,2,j)
  sns.distplot(wine_dataset[i],hist=True)
  j+=1
```

```python
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
j=1
col=["price","points"]
for i in col:
  plt.subplot(2,2,j)
  sns.countplot(x=wine_dataset.iloc[:50][i])
  j+=1
```
